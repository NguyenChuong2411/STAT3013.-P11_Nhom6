# Introduction
In a world where paper money matters, gold is the reserve of any country. In the case of inflation, it can be used as an alternative replacement for the money we are used to. If ordinary money becomes worthless in the process of inflation, the country's gold reserve will be released to support the economy. This system was developed taking into account the fact that paper money can easily devalue, and gold throughout it's history has not lost in value until now.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the monthly data and ensure 'Date' is in datetime format
month = pd.read_csv('GoldPrices_monthly_1992_2020.csv')
month['Date'] = pd.to_datetime(month['Date'], format='%Y-%m')

# Set up the plot
plt.figure(figsize=(16, 10), facecolor='#FCFDFE')
sns.set_style("white")
plt.title('Dynamics of Gold Price', fontname='serif', fontsize=30, y=1.05, color='#4F6457')

# Plot the monthly gold price data
a = sns.lineplot(data=month, x='Date', y='Gold_Price', color='#C99E10', linewidth=3)
plt.grid(color='gray', linestyle=':', axis='y', alpha=0.5, zorder=0, dashes=(1,7))
a.patch.set_facecolor('#FCFDFE')
plt.xlabel('Year', fontname='serif', fontsize=15, labelpad=20)
plt.ylabel('Price, $US', fontname='serif', fontsize=15, labelpad=20)

# Set the x-axis to display only years
a.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y'))
plt.xticks(ha='right', fontname='serif', fontsize=11.5)

# Define events with specific dates, labels, and positions
events = [
    {'date': '2001-09', 'price': 580, 'ymin': 0.06, 'ymax': 0.23, 'text_x': '1999-01', 'text_y': 610, 
     'label': 'Terrorist attacks\non the United States'},
    {'date': '2003-03', 'price': 700, 'ymin': 0.1, 'ymax': 0.3, 'text_x': '2003-01', 'text_y': 750, 
     'label': 'The Iraq War'},
    {'date': '2008-09', 'price': 1120, 'ymin': 0.37, 'ymax': 0.53, 'text_x': '2006-06', 'text_y': 1135, 
     'label': 'Global financial\ncrisis'},
    {'date': '2011-08', 'price': 1100, 'ymin': 0.54, 'ymax': 0.92, 'text_x': '2013-01', 'text_y': 1000, 
     'label': 'Sovereign Debt Crisis\nand U.S. Credit Downgrade'},
    {'date': '2015-12', 'price': 1400, 'ymin': 0.52, 'ymax': 0.7, 'text_x': '2016-06', 'text_y': 1500, 
     'label': "FED's First Interest\nRate Hike since\nthe 2008 Financial Crisis"}
]

# Convert event dates and text positions to datetime
for event in events:
    event_date = pd.to_datetime(event['date'])
    text_date = pd.to_datetime(event['text_x'])
    price_event = event['price']
    
    # Draw vertical line from ymin to ymax for the event date
    a.axvline(x=event_date, ymin=event['ymin'], ymax=event['ymax'], linestyle='-', color='#4F6457', linewidth=2.2)
    
    # Plot the event marker at the top of the line
    a.plot(event_date, price_event, 'ko', markersize=7, fillstyle='full', color='#4F6457', markeredgewidth=3)
    
    # Place the label text slightly above or beside the marker
    a.text(text_date, event['text_y'], event['label'], ha='center', fontsize=12, fontname='serif', color='#4F6457')

# Customize plot spines
for spine in ['right', 'top']:
    a.spines[spine].set_visible(False)
for spine in ['left', 'bottom']:
    a.spines[spine].set_linewidth(1.3)
    a.spines[spine].set_color('#9A9EAB')

plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly as py
import plotly.graph_objs as go
import plotly.express as px
from plotly.offline import init_notebook_mode
init_notebook_mode(connected = True)
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.max_columns', None)

#########################################################

GDP_year = pd.read_csv('Dataset.csv')
```

```python
GDP_year['Date'] = GDP_year['Date'].apply(lambda x: x[:-3]).astype('int')
```

```python
# Filter for rows where the year is a multiple of 10
half_decade = year.query("Date % 5 == 0").copy()  # Use .copy() to avoid setting on a slice
half_decade['Dynamics'] = 0

# Calculate half_decade-over-half_decade percentage change
for i in range(1, len(half_decade)):
    half_decade.at[half_decade.index[i], 'Dynamics'] = round(((half_decade.iloc[i, 1] * 100) / half_decade.iloc[i - 1, 1]) - 100, 2)

half_decade
```

```python
# Define rotation for each annotation
rotation_dict = {
    1: -20,  # Rotation for 1995-2000
    2: 20,   # Rotation for 2000-2005
    3: 55,   # Rotation for 2005-2010
    4: 20,   # Rotation for 2010-2015
    5: 40    # Rotation for 2015-2020
}

# Set up the plot
plt.figure(figsize=(16, 10), facecolor='#FCFDFE')
sns.set_style("white")
plt.title('Dynamics of the Gold Price by Half - Decade', fontname='serif', fontsize=30, y=1.05, color='#4F6457')

# Line and scatter plots
a = sns.lineplot(data=half_decade, x='Date', y='Gold_Price', color='#C99E10', linewidth=3)
sns.scatterplot(data=half_decade, x='Date', y='Gold_Price', color='#C99E10', s=90)
a.patch.set_facecolor('#FCFDFE')
plt.grid(color='gray', linestyle=':', axis='x', alpha=0.5, zorder=0, dashes=(1,7))

# Labels and ticks
plt.xlabel('Year', fontname='serif', fontsize=15, labelpad=20)
plt.ylabel('Price, $US', fontname='serif', fontsize=15, labelpad=20)
plt.xticks(fontname='serif', fontsize=11)
plt.yticks(fontname='serif', fontsize=11)

# Annotate percentage changes in the middle between each point
for i in range(1, len(half_decade)):
    x_mid = (half_decade['Date'].iloc[i] + half_decade['Date'].iloc[i - 1]) / 2
    y_mid = (half_decade['Gold_Price'].iloc[i] + half_decade['Gold_Price'].iloc[i - 1]) / 2
    percentage_change = half_decade['Dynamics'].iloc[i]
    color = '#265C00' if percentage_change > 0 else '#8D230F'
    rotation = rotation_dict.get(i, 0)  # Use rotation from the dictionary, default to 0 if not specified
    
    # Add text annotation with manual rotation and centered alignment
    a.text(x_mid, y_mid, f'{percentage_change:+.2f}%', 
           ha='center', fontsize=14, fontname='serif', color=color, 
           rotation=rotation)

# Style adjustments
for spine in ['right', 'top']:
    a.spines[spine].set_visible(False)
for spine in ['left', 'bottom']:
    a.spines[spine].set_linewidth(1.3)
    a.spines[spine].set_color('#9A9EAB')

plt.show()
```
