```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load dataset and preprocessing
data = pd.read_csv('Dataset.csv')  
data['Date'] = pd.to_datetime(data['Date'])
scaler = MinMaxScaler()
columns_to_scale = ['Gold_Price', 'USD_Price', 'GDP', 'Brent']
data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])

# Set y-axis tick positions
date_ticks = pd.date_range(start=data['Date'].min(), end=data['Date'].max(), freq='2Y')

# Plot scaled Gold Price and USD Index
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Gold_Price'], label='Gold Price', color='gold', alpha=0.8)
plt.plot(data['Date'], data['USD_Price'], label='USD Index', color='red', alpha=0.8)
plt.title('Scaled Gold Price and USD Index Over Time')
plt.xlabel('Date')
plt.ylabel('Scaled Value')
plt.xticks(date_ticks, date_ticks.strftime('%Y'))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('gold_usd_chart.png') 
plt.show()

# Plot scaled Gold Price and GDP
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Gold_Price'], label='Gold Price', color='gold', alpha=0.8)
plt.plot(data['Date'], data['GDP'], label='GDP', color='blue', alpha=0.8)
plt.title('Scaled Gold Price and GDP Over Time')
plt.xlabel('Date')
plt.ylabel('Scaled Value')
plt.xticks(date_ticks, date_ticks.strftime('%Y'))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('gold_gdp_chart.png') 
plt.show()

# Plot scaled Gold Price and Brent
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Gold_Price'], label='Gold Price', color='gold', alpha=0.8)
plt.plot(data['Date'], data['Brent'], label='Brent Crude Oil Prices', color='black', alpha=0.8)
plt.title('Scaled Gold Price and Brent Crude Oil Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Scaled Value')
plt.xticks(date_ticks, date_ticks.strftime('%Y'))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('gold_brent_chart.png')  
plt.show()
```
